```python
!pip install selenium
!apt-get update
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin

import pandas as pd 
import sys
from selenium import webdriver
from selenium.webdriver.common.by import By

sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
# Entro en la página que contiene la info que quiero scrapear
url = 'https://app.tikr.com/stock/financials?cid=29096&tid=11311662&ref=nf4x8h'
driver.get(url)
elem = driver.find_element(by=By.XPATH, value="//body")
source_code = elem.get_attribute("outerHTML")

driver.maximize_window()
```

```python
headers = driver.find_elements(by=By.XPATH, value='/html/body/div/div/div/div[1]/main/div/div/div[2]/div/div/div[3]/div/div/div/div[2]/div[1]/div/table/thead/tr/th')
print("Headers: \n")
for i in range(len(headers)):
  print(headers[i].text)
```

### Scraping de los Datos Financieros de Google


```python
# Con un bucle recorro el header del futuro df para transformarlo 
# de objeto de selenium a texto

data_names= driver.find_elements(by=By.XPATH, value='/html/body/div/div/div/div[1]/main/div/div/div[2]/div/div/div[3]/div/div/div/div[2]/div[1]/div/table/tbody/tr/td[1]')
for i in range(len(data_names)):
  data_names[i]=data_names[i].text
print(data_names)
```

```python
y2016= driver.find_elements(by=By.XPATH, value='/html/body/div/div/div/div[1]/main/div/div/div[2]/div/div/div[3]/div/div/div/div[2]/div[1]/div/table/tbody/tr/td[2]')
for i in range(len(y2016)):
  y2016[i]=y2016[i].text
print("Data 2016: ", y2016)
y2017= driver.find_elements(by=By.XPATH, value='/html/body/div/div/div/div[1]/main/div/div/div[2]/div/div/div[3]/div/div/div/div[2]/div[1]/div/table/tbody/tr/td[3]')
for i in range(len(y2017)):
  y2017[i]=y2017[i].text
print("Data 2017: ",y2017)
y2018= driver.find_elements(by=By.XPATH, value='/html/body/div/div/div/div[1]/main/div/div/div[2]/div/div/div[3]/div/div/div/div[2]/div[1]/div/table/tbody/tr/td[4]')
for i in range(len(y2018)):
  y2018[i]=y2018[i].text
print("Data 2018: ",y2018)
y2019= driver.find_elements(by=By.XPATH, value='/html/body/div/div/div/div[1]/main/div/div/div[2]/div/div/div[3]/div/div/div/div[2]/div[1]/div/table/tbody/tr/td[5]')
for i in range(len(y2019)):
  y2019[i]=y2019[i].text
print("Data 2019: ",y2019)
y2020= driver.find_elements(by=By.XPATH, value='/html/body/div/div/div/div[1]/main/div/div/div[2]/div/div/div[3]/div/div/div/div[2]/div[1]/div/table/tbody/tr/td[6]')
for i in range(len(y2020)):
  y2020[i]=y2020[i].text
print("Data 2020: ",y2020)
y2021= driver.find_elements(by=By.XPATH, value='/html/body/div/div/div/div[1]/main/div/div/div[2]/div/div/div[3]/div/div/div/div[2]/div[1]/div/table/tbody/tr/td[7]')
for i in range(len(y2021)):
  y2021[i]=y2021[i].text
print("Data 2021: ",y2021)
```

```python
print("Names ",len(data_names))
print("2016 ",len(y2016))
print("2017 ",len(y2017))
print("2018 ",len(y2018))
print("2019 ",len(y2019))
print("2020 ",len(y2020))
print("2021 ",len(y2021))
```

**`Supplementary Data es un valor que aparece en data_names pero no tiene ningún valor asociado a ninguna columna, por ello se borra. Una vez que todas las listas tienen el mismo número de valores se podrá construir el dataframe`**

```python
index = data_names.index('Supplementary Data:')
data_names.pop(index) 
```

```python
# Compruebo de nuevo que todas las listas tienen el mismo tamaño
print("Names ",len(data_names))
print("2016 ",len(y2016))
print("2017 ",len(y2017))
print("2018 ",len(y2018))
print("2019 ",len(y2019))
print("2020 ",len(y2020))
print("2021 ",len(y2021))
```

```python
googl_income_df = pd.DataFrame(columns=['Income Statement', '31/12/16', '31/12/17', '31/12/18',
                                     '31/12/19', '31/12/20', '31/12/21'])
for i in range(len(data_names)):
    googl_income_df = googl_income_df.append({'Income Statement':data_names[i], '31/12/16':y2016[i], '31/12/17':y2017[i], '31/12/18':y2018[i],
                                     '31/12/19':y2019[i], '31/12/20':y2020[i], '31/12/21':y2021[i]}, ignore_index=True)
googl_income_df
```

```python
# Los datos de 2016 requieren una suscripción a la plataforma. 
# Elimino la columna para no tener datos vacíos.

googl_income_df=googl_income_df.drop("31/12/16",axis=1)
```

```python
googl_income_df.set_index('Income Statement')
```

### Exportando los datos en distintos formatos

```python
# excel
googl_income_df.to_excel('GOOGL_Income_Statement.xlsx', index=False)
```
